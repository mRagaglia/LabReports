```python
# plots will be shown inline 
%matplotlib inline 
import matplotlib.pyplot as plt

import numpy
from numpy import sqrt,floor

import numpy as np
import scipy as sp
from scipy import stats


import pandas as pd
import random
import math
from matplotlib.ticker import ScalarFormatter
```

```python
y1=np.asarray([68.54,134.93,198.94,260.41,331.77])
x1=np.asarray([50,100,150,200,255])

y1=y1*5/1024

sum1=x1.var()/len(x1)

xy=(x1*y1).mean()
x=x1.mean()
y=y1.mean()
x2=(x1**2).mean()

m1=(xy-x*y)/(x2-x**2)
c1=(x2*y-x*xy)/(x2-x**2)
merr1=1/((x2-x**2)*sum1)
cerr1=merr1*x2
cov=-x*merr1
p=-x/x2**0.5
merr1=(merr1**0.5)
cerr1=cerr1**0.5
print('----parametri retta----')
print(1/m1,c1,merr1,cerr1,cov,p)

#plot punti con errori

fig, ax = plt.subplots(figsize=(10, 6))

ax.errorbar(x1,y1 , linestyle='none',label='Punti sperimentali', markeredgecolor='black', alpha=0.5,marker='o',color='Blue',  markersize=6 ,ecolor='black')
ax.grid(linestyle='--')


#plot fit

x=np.linspace(0,300)
y1=m1*x+c1

ax.plot(x,y1, ls='-.',color='orangered', label='Retta di fit con m=4.98')
ax.set_title('Fit per la risposta lineare del lock-in', fontsize=15)
ax.set_xlim(0,300)
ax.set_ylim(0.2510)

#caratteritiche grafico
ax.set_ylabel('Ampiezza $V_{out}$ [mV]', fontsize=14)
ax.set_xlabel('Ampiezza $V_{in}$ [mV]', fontsize=14)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.0), fancybox=True, shadow=True, ncol=2, fontsize=11, facecolor='white')
plt.show()
```
